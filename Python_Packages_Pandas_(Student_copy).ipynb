**Pandas** - Contains data structures and data manipulation tools designed to make data cleansing and analysis fast and easy

Adopts coding style from Numpy but main difference is that Pandas can also handle heterogeneous data.

There are 2 primary Pandas data structures - 
1. **Series** - A one-dimensional object, similar to an array, list, or column in table. Each item in a Series, however, is assigned to an entry in an index. It is a 1D labeled homogeneously-typed array
2. **DataFrame** - Represents a rectangular table of data (2D) and contains an ordered collection of columns, each of which can be a different value type (numeric, string, boolean, etc.). Has both a row and column index.

```
import pandas as pd
```

### Series

```
# series
series1 = pd.Series([4, 7, -5, 3])

print("Series:\n", series1)
print("Series values:\n", series1.values)
print("Series index:\n", series1.index)
```

```
series2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"]) 

print("Series:\n", series2)
print("Series values:\n", series2.values)
print("Series index:\n", series2.index)
```

```
# element access

series2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"]) 

print("Element at index d", series2["d"])

series2["d"] = 6
print("Element at index d", series2["d"])

print(series2)

print(series2[["a", "c", "d"]]) # why do you need the double square brackets?
print(series2[[2, 3, 0]])
```

```
import numpy as np

series2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"]) 

print("NumPy like filtering:\n", series2[series2 > 0])
print("NumPy like arithmetic:\n", series2 * 2)

# you can also pass series to numpy functions
print(np.absolute(series2))
```

```
series2 = pd.Series([4, 7, -5, 3], index=["d", "b", "a", "c"]) 

series2.name = "Numbers"
series2.index.name = "Letters"

print("Value title: ", series2.name)
print("Index title: ", series2.index.name)
print(series2)
```

```
# use a dictionary to create series

data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data)
print(s)
```

```
 
data = {'a' : 0., 'b' : 1., 'c' : 2.}
s = pd.Series(data,index=['b','c','d','a'])
print(s) 
 
# Index order is persisted and the missing element is filled with NaN (Not a Number).

```

```
# create a series 
data = np.array(['g','e','e','k','s','f', 'o','r','g','e','e','k','s'])
ser = pd.Series(data)
  
# retrieve the first element
print(ser[:5])
```

### Series Indexing

```
obj = pd.Series(np.arange(4.), index=["a", "b", "c", "d"])
print(obj)

print(obj["b"])

print(obj[1])

print(obj[[2,3]])

print(obj[2:4])

print(obj[["b", "c"]])

print(obj[obj<2])

print(obj["c":"d"])

print(obj["d":"b":-1])
```

### Sorting

```
series = pd.Series(range(4), index=["d", "a", "b", "c"])

print(series)
print(series.sort_index())
print(series.sort_index(ascending=False))          
```

### DataFrame

Dataframes can be created from lists, dictionaries, dataframes, series, 2D arrays

You can optionally specify row labels and column labels
If no labels are set, then the rows and columns are labeled 0,1,â€¦

We generally load data from csv files into dataframes.


```
# creating dataframe from dictionary
data = {"state": ["Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year": [2001, 2002, 2001, 2002, 2003],
        "population": [1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data)

print(frame)
print(frame.head()) # prints first 5 rows -- default
print(frame.head(2))
```

```
data = {"state": ["Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
        "year": [2001, 2002, 2001, 2002, 2003],
        "population": [1.7, 3.6, 2.4, 2.9, 3.2]}

frame = pd.DataFrame(data, columns=["year", "state", "population"], index=["one", "two", "three", "four", "five"])
print(frame)
```

```
# accessing elements

# retrieve columns

print(frame["state"])

print(frame[["state", "population"]])

print(frame.year)
```

```
# add a new column which returns true if state is Ohio

frame["new_col"] = frame["state"] == "Ohio"
print(frame)

# to delete column

# del frame["new_column"]
```

```
# create a 2D array from dataframe

arr_2d = frame.values
print(arr_2d)
```

### DataFrame indexing

```
data = pd.DataFrame(np.arange(12).reshape(3, 4), index=["Colorado", "Utah", "New York"], columns=["a", "b", "c", "d"])

print(data)
print(data[["b", "c"]])
print(data[data["c"]>5])
```

```
#select rows by multiple column values

data = {"name": ["John", "Mary", "Peter", "Jeff", "Bill", "Lisa"],
        "state": ["california", "dc", "california", "texas", "washington", "california"],
        "num_children": [2, 2, 0, 1, 2, 1],
        "num_pets": [0, 4, 0, 5, 0, 0]}

frame = pd.DataFrame(data, columns=["name", "state", "num_children", "num_pets"])
print(frame)

#select people who are in california state and number of theit children is greater than number of their pets

frame[(frame["state"]== "california") & ( frame["num_children"]> frame["num_pets"])]
```

```
print(data.iloc[2]) #integer location
print(data.iloc[2, [3, 0, 1]])
print(data.iloc[[1,2], [3, 0, 1]])

# iloc is used for integer based indexing
# Use : to display all rows/columns
# Data.iloc[:,:] 

print(data.iloc[:,:])
```

```
# loc function - loc is label based location index
# Used for label based indexing

print(data.loc["Colorado"])
print(data.loc["Colorado", ["b", "c"]])
print(data.loc[["Colorado", "Utah"], ["b", "c"]])
print(data.loc["Colorado":"Utah", "b":"c"])
```

### Pandas sorting

```
frame = pd.DataFrame({"b": [4, 7, -3, 2], "a": [0, 1, 0, 1]})
print(frame)

print(frame.sort_values(by=["a", "b"], ascending=False))
print(frame.sort_values(by=["a", "b"], ascending=[True, False]))
```
