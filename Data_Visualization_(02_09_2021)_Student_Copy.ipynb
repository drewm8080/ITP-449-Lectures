### Line Plot

```
year = [2016, 2017, 2018, 2019, 2020]
enrollment = [70, 150, 130, 180, 220]
plt.plot(year, enrollment, marker='o')
plt.show()
```

```
# Change the marker

plt.plot(year, enrollment, marker ='o')
plt.show()

# https://matplotlib.org/2.0.2/api/markers_api.html

# Change the line style

plt.plot(year, 	enrollment, marker ='o', linestyle='--')
plt.show()

# Set the color

plt.plot(year, enrollment, 	marker ='o', linestyle='--', color = 'r')
plt.show()

# https://matplotlib.org/2.0.2/api/colors_api.html

# Simpler usage

plt.plot(year, enrollment, 'ro--')
plt.show()

# To see the changes separately, you have to use plt.show() after every plt.plot(). Otherwise, the changes get applied on the same plot.
```

```
# Add labels, title and controlling axes ranges
plt.plot(year, enrollment, 'ro--')

# Add labels
plt.xlabel('Year')
plt.ylabel('Enrollment')

# Add title
plt.title('Class Enrollment')

# Control axes range
# syntax - plt.axis([xmin, xmax, ymin, ymax])
plt.axis([2014, 2022, 0, 300])

plt.show()
```

```
year = [2016, 2017, 2018, 2019, 2020]
ITP449 = [70, 150, 130, 180, 220]
ITP115 = [180, 250, 390, 540, 720]

plt.plot(year, ITP449, marker ='o', linestyle='--', color = 'r', label='ITP449')
plt.plot(year, ITP115, marker ='s', linestyle='-.', color = 'b', label='ITP115')

# Add labels
plt.xlabel('Year')
plt.ylabel('Enrollment')

# Add title
plt.title('ITP Enrollment')

# Control axes range
plt.axis([2014, 2022, 0, 800])

# legend and grid
plt.legend()
plt.grid(True, color='y', linestyle=':')

plt.show()
```

### Pie Charts

```
plt.style.use('ggplot')
browsers = ['Chrome', 'Firefox', 'Internet Explorer', 'Safari', 'Others']
marketShare = [61.64, 18.98, 11.02, 4.23, 4.13]
plt.pie(marketShare, labels=browsers, autopct='%.1f%%', startangle=45)
plt.title('Browser Market Share')
plt.axis('equal')
plt.show()
```

```
plt.style.use('seaborn')
labels = ['Chrome', 'Firefox', 'Internet Explorer', 'Safari', 'Others']
marketShare = [61.64, 18.98, 11.02, 4.23, 4.13]
plt.pie(marketShare, labels=labels, autopct='%.1f%%', startangle=45,
        explode=[0, 0, 0, 0.5, 0.5]) # split and display portion of pie chart
plt.title('Browser Market Share')
plt.axis('equal')
plt.legend(labels, loc='lower left') # legend
plt.show()
```

### Figures

When you create a plot, a figure object is automatically created.
You can create multiple figures and refer to them by their number.


```
plt.style.use('ggplot')

plt.figure(1)
x = np.arange(-4, 4.1, 0.1)
plt.plot(x, x**2)
plt.axis([-4, 4, -70, 70])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('x^2')

plt.figure(2)
x = np.arange(-4, 4.1, 0.1)
plt.plot(x, x**4)
plt.axis([-4, 4, -70, 70])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('x^4')

plt.show()
```

### Subplots - A set of plots within a figure. 

Each subplot focuses on a slice of the overall data.
Each subplot has an index so that each plot can be addressed - 

plt.subplot(number of x plots, number of y plots, chart #)

example - 

plt.subplot(2, 2, 1)
plt.subplot(2, 2, 2)
plt.subplot(2, 2, 3)
plt.subplot(2, 2, 4)

```
plt.style.use('ggplot')

x = np.arange(0, 4, 0.1)

plt.subplot(1, 3, 1) # 1 row, 2 cols, chart 1
plt.plot(x, x**(1/2), 'y^', x, x**(1/3), 'bo', x, x**(1/4), 'r--')
plt.axis([0, 4, 0, 3])
plt.legend(['x^(1/2)', 'x^(1/3)', 'x^(1/4)'])

plt.subplot(1, 3, 2) # 1 row, 2 cols, chart 2
plt.plot(x, x**2, 'y^', x, x**3, 'bo', x, x**4, 'r--')
plt.axis([0, 4, 0, 70])
plt.legend(['x^2', 'x^3', 'x^4'])

plt.subplot(1, 3, 3)

plt.show()
```

You can create a blank figure
And then add subplots to it

```
# clear runtime to get rid of grids
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)

plt.show()
```

```
myFig = plt.figure()

ax1 = myFig.add_subplot(2, 2, 1)
ax2 = myFig.add_subplot(2, 2, 2)
ax3 = myFig.add_subplot(2, 2, 3)

ax1.hist(np.random.randn(1000), bins=50, color='m')
ax2.scatter(np.arange(30), np.arange(30) + 2 * np.random.randn(30), color='b')
ax3.plot(np.random.randn(50).cumsum(), 'g--') # cumulation sum of standard normal distribution of 50 numbers (mean = 0, variance = 1)
plt.show()
```

```
# Generates an array of subplots 2 x 2
# myFig refers to figure
# myAxes refers to axes
# myAxes[i, j] refers to a subplot

myFig, myAxes = plt.subplots(2, 2)
plt.show()
```

```
# histogram

myFig, myAxes = plt.subplots(2, 2, sharex=True, sharey=True) # share x and y axis
for i in range(2):
    for j in range(2):
        myAxes[i, j].hist(np.random.randn(1000), bins=50, color='k', alpha=0.5)
plt.show()
```

```
myFig, myAxes = plt.subplots(1, 4, sharey=True, squeeze=False) # Squeeze = False is used to allow proper indexing of axes

myData = np.random.randn(1000)
myAxes[0, 0].hist(myData, bins=10, color='r')
myAxes[0, 1].hist(myData, bins=20, color='b')
myAxes[0, 2].hist(myData, bins=50, color='k')
myAxes[0, 3].hist(myData, bins=100, color='g')
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())
plt.show()
```

```
# setting ticks and labels and rotating tick labels

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum())
ticks = ax.set_xticks([0, 250, 500, 750, 1000]) # xticks
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small') # xtick labels
ax.set_title('Plot using subplots')
ax.set_xlabel('Stages')
plt.show()
```

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.plot(np.random.randn(1000).cumsum(), 'k', label='one')
ax.plot(np.random.randn(1000).cumsum(), 'k--', label='two')
ax.plot(np.random.randn(1000).cumsum(), 'k.', label='three')
ax.legend(loc='best') # legend position
plt.show()
```

```
import seaborn as sb
```

```
# load csv data

puzzleCompletionTime = pd.read_csv('grade5ScienceFairProject.csv')
```

```
# swarm plot

sb.set_style('whitegrid')
sb.swarmplot(y='Seconds', data=puzzleCompletionTime)
plt.show()
```

```
sb.set_style('whitegrid')
sb.swarmplot(x='Music', y='Seconds', data=puzzleCompletionTime, size=10)
plt.show()
```

```
# superimpose a box plot on a swarm plot

sb.set_style('whitegrid')
sb.swarmplot(x='Music', y='Seconds', data=puzzleCompletionTime, size=10)
sb.boxplot(x='Music', y='Seconds', data=puzzleCompletionTime)
plt.show()
```
