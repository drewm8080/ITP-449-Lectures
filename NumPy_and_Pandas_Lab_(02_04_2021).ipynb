### **1. Split a (8, 4) shaped matrix of random integers, ranging from 10 to 20, into 4 arrays with shapes (4, 2)**


```
m = np.random.randint(low = 10, high = 21, size = (8,4))
```

```
print(m)
half = np.hsplit(m,2)
sub_matrices = []
for e in half:
  sub_matrices.append(np.vsplit(e,2))

print(sub_matrices)
# [left,right] = np.hsplit(m,2)
# [ul,ll] = np.vsplit(left,2)
# [ur,lr] = np.vsplit(right,2)
# print(ur)
# print(lr)
```

```
# Hint: You can use np.hsplit and np.vsplit

# Type solution below

x = np.random.randint(low=10, high=20, size=(8, 4))

[x_left, x_right] = np.hsplit(x, 2)
print(x_left)

[x_left_top, x_left_bottom] = np.vsplit(x_left, 2)
[x_right_top, x_right_bottom] = np.vsplit(x_right, 2)

print(x, "\n\n")
print(x_left_top, "\n\n")
print(x_left_bottom, "\n\n")
print(x_right_top, "\n\n")
print(x_right_bottom, "\n\n")
```
